```python
import matplotlib.pyplot as plt
import pandas as pd
import pickle, os
from matplotlib.pyplot import figure
```

```python
data_path = "data/evaluation/"
```

```python
pickles = []
configs = []
for config in os.listdir(data_path):
    configs.append(config)
    result_path = os.path.join(data_path, config).replace('\\', '/')
    result_path = os.path.join(result_path, "result.pkl").replace('\\', '/')
    file = open(result_path, "rb")
    pickles.append(pickle.load(file))
pickles[0]
```

```python
len(pickles)
```

```python
#create precision recall curve
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
for i in range(0, 5):
    recall = pickles[i]["recall"]
    precision = pickles[i]["precision"]
    ax.plot(recall, precision, label = configs[i])
ax.legend()

#add axis labels to plot
ax.set_title('Precision-Recall Curve')
ax.set_ylabel('Precision')
ax.set_xlabel('Recall')

#display plot
plt.show()
```
